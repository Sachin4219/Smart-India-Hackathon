```python
casetype_score={
'Murder and Homicide Cases':39,
'Theft and Robbery Cases':38,
'Fraud and Financial Crimes':37,
'Assault and Battery Cases':36,
'Cybercrime Cases':35,
'Domestic Violence Cases':34,
'Child Custody Cases':33,
'Adoption Cases':32,
'Guardianship Cases':31,
'Divorce Cases':30,
'Matrimonial Disputes':29,
'Maintenance Cases':28,
'Restitution Cases':27,
'Succession Cases':26,
'Fundamental Rights Cases':25,
'Public Interest Litigation (PIL)':24,
'Pollution Control Cases':23,
'Conservation Cases':22,
'Custodial Death Cases':21,
'Illegal Detention Cases':20,
'Discrimination Cases':19,
'Bankruptcy and Insolvency Cases':18,
'Contractual Disputes':17,
'Intellectual Property Disputes':16,
'Income Tax Cases':15,
'Goods and Services Tax (GST) Cases':14,
'Money Suits':13,
'Property Suits':12,
'Commercial Suits':11,
'Rent Suits':10,
'Injunction Suits':9,
'Specific Performance Suits':8,
'Partition Suits':7,
'Declaration Suits':6,
'Intellectual Property Cases':5,
'Consumer Cases':4,
'Property Cases':3,
'Labor and Employment Cases':2,
'Writ Petitions':1
}
```

```python
df.columns
```

```python
df['score']=0.0
```

```python
for i in range(df.shape[0]):
    df['score'][i] += casetype_score[df['Case Type'].iloc[i]]
    df['score'][i] += df['Number of parties involved in the case'].iloc[i]
    if df['Diversity of Interests between parties'].iloc[i] == 'Yes':
        df['score'][i] += 1
    if df['Presence of Counter Claims'].iloc[i] == 'Yes':
        df['score'][i] += 1
    df['score'][i] += df['Party or Member of Party Deceased'].iloc[i]
    df['score'][i] += df['Time for evidence study'].iloc[i]
    df['score'][i] += df['Critical Evidence related to human rights violation or public safety'].iloc[i]
    if df['Urgency by Parties'].iloc[i] == 'Yes':
        df['score'][i] += 1
    df['score'][i] += df['Amount of Financial Claims made in the case'].iloc[i] / 10000
    if df['Vulnerable Population (Children/ Elderly) involved in the case'].iloc[i] == 'Yes':
        df['score'][i] += 1
    if df['Influential Person Involved (Businessman/Politician/Actor etc.)'].iloc[i] == 'Yes':
        df['score'][i] += 1
    df['score'][i] += df['Public Interest'].iloc[i] / 10
    if df['Overall Public Sentiment'].iloc[i] == 'Negative':
        df['score'][i] += df['Public Interest'].iloc[i] / 100
    df['score'][i] += df['Number of witnesses'].iloc[i]
    df['score'][i] += df['Number of hearings held'].iloc[i]
    print(i)
```

```python
df['priority_float'] = df['score']*100/df['score'].max()
```

```python
df['priority_float'].min()
```

```python
df['priority_bucket']=np.floor(df['priority_float'])
```
